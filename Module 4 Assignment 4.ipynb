```python
#allow output from every line
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
```

```python
# 2 - Choose a league, season, and 4 teams within the league. For the given season, collect the roster
#table from sports reference for ALL 4 teams. This should result in you having 4 different CSVs
#(one for each team).

    #I decided to select the Detroit Red Wings, Boston Bruins, Philadelphia Flyers, & Tampa Bay Lightning from
    # 23-24 NHL season.

# 3 - In your notebook file, read all 4 CSVs in, with the index being the player name.

red_wings_roster = pd.read_csv("Red_Wings_Roster.csv", index_col='Player')
bruins_roster = pd.read_csv("Bruins_Roster.csv", index_col='Player')
flyers_roster = pd.read_csv("Flyers_Roster.csv", index_col='Player')
lightning_roster = pd.read_csv("Lightning_Roster.csv", index_col='Player')
```

```python
# 4 - Concatenate the 4 tables into 1 all-inclusive table.

four_hockey_rosters = pd.concat([red_wings_roster, bruins_roster, flyers_roster, lightning_roster], ignore_index=False)
four_hockey_rosters
```

```python
four_hockey_rosters.to_csv("DET_BOS_PHI_TAM_Rosters.csv")
```
